```python
import actonet
import biolqm
from colomoto.minibn import BooleanNetwork
from colomoto_jupyter import tabulate
import cabean # for attractor computation
```

In this notebook we compare the results obtained with actonet, stable motifs and caspo-control on the tumor invasion model

Definition of targets 

```python
targets = {"CellCycleArrest": 1,"EMT":1,"Invasion":1,"Migration":1,"Metastasis":1,"Apoptosis":0}
```

Computation of wild-type model attractors

```python
model = biolqm.load("http://ginsim.org/sites/default/files/SuppMat_Model_Master_Model.zginml")
test = biolqm.to_minibn(model, ensure_boolean=True)
att=cabean.attractors(test)
tabulate(att)
```

# with sources 

```python
sources = {"ECMicroenv":1,"DNAdamage":0}
```

## actonet solution
The solution set computed by actonet for the source,target couple defined previsouly is : an={}
The model with the sources fixed has a single stable state which verifies the desired values for the targets :

```python
test1 = biolqm.to_minibn(model, ensure_boolean=True)
for n, v in sources.items():
    test1[n] = v
att1=cabean.attractors(test1)
tabulate(att1)
```

This unique stable state is a stable state of the wild-type model : 

```python
t=[a==list(att1)[0] for a in list(att)]
t
```

## stable motifs solutions
The solution set computed by stable motifs for the source,target couple defined previsouly is :sm={(SNAI2=1),(p53=0)}
The model with the sources and solution fixed has a single stable state which verifies the desired values for the targets :

```python
test2 = biolqm.to_minibn(model, ensure_boolean=True)
for n, v in sources.items():
    test2[n] = v
    test2["SNAI2"] = 1
att2=cabean.attractors(test2)
tabulate(att2)
```

This unique stable state is a stable state of the wild-type model : 

```python
t=[a==list(att2)[0] for a in list(att)]
t
```

```python
test3 = biolqm.to_minibn(model, ensure_boolean=True)
for n, v in sources.items():
    test3[n] = v
    test3["p53"] = 0
att3=cabean.attractors(test3)
tabulate(att3)
```

```python
t=[a==list(att3)[0] for a in list(att)]
t
```
